### Task 0
Load the “/dbfs/databricks-datasets/flights/departuredelays.csv” file. Convert date to the type of datetime. All the flights are recorded in the year of 2014 between January and March. If you see a date as 1020605, it means that the flight is scheduled at 6:05AM on January 2nd.

```
import pandas as pd 
import numpy as np
import matplotlib

#df = pd.read_csv(“/dbfs/databricks-datasets/flights/departuredelays.csv”)

df = spark.read\
  .format("csv")\
  .option("header", "true")\
  .option("inferSchema", "true")\
  .load("/databricks-datasets/flights/departuredelays.csv")\
  .toPandas()

display(df)
```

```
df['date'] = df['date'].astype('str')
df['date'] = df['date'].apply(lambda x: '20140'+x)

df['date'] = pd.to_datetime(df['date'], format = '%Y%m%d%H%M')

display(df)
```

### Task 1 
Find the top 5 busiest airports (the business of an airport is defined as the total number of departure and arrival flights). Store these
airports in a list and print the list.

```
busiest_origin = df.groupby(['origin'])['origin'].count().reset_index(name='total_origin')
display(busiest_origin)
```

```
origin_airports = busiest_origin.rename(columns={'origin':'airport'})
display(origin_airports)
```

```
#x = busiest_origin.sort_values(by='total_origin', ascending=False)
#display(x.head(5))
```

```
busiest_dest = df.groupby(['destination'])['destination'].count().reset_index(name='total_dest')
display(busiest_dest)
```

```
dest_airports = busiest_dest.rename(columns={'destination':'airport'})
display(dest_airports)
```

```
merged = origin_airports.merge(dest_airports, on='airport')
#total_busiest = busiest_origin.append(busiest_dest)
display(merged)
```

```
#df_merged = pd.concat([busiest_origin, busiest_dest], axis=1)
#display(df_merged)
#merged['total'] = ['total_origin'] + ['total_dest']
merged['total'] = merged.apply(lambda x: x['total_origin'] + x['total_dest'], axis=1)
display(merged)
```

```
top5 = merged.sort_values(by='total', ascending=False)
display(top5[['airport', 'total']].head(5))
```
